# EDA

# Number of attack per year

```python
x_year = data['iyear'].unique()
y_count_year = data['iyear'].value_counts(dropna=False).sort_index()
plt.figure(figsize=(18,10))
sns.barplot(x=x_year,y=y_count_year,palette='rocket')
plt.xticks(rotation = 45)
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks per year')
plt.title('Attack_of_years',fontsize=16)
plt.show()
```

# Incident count per month(overall)

```python
plt.figure(figsize=(15,7))
sns.countplot(data=df,x='imonth')

```

# Terrorist activities by region each year

```python
pd.crosstab(df.iyear,df.region_txt).plot(kind='area',figsize=(15,6))
plt.title('Terrorist Activities by Region in each year')
plt.ylabel('Number of Attacks')
plt.show()
```

# Top countries affected by terrorist attack

```python
plt.subplots(figsize=(15,6))
plt.bar(df['country_txt'].value_counts()[:15].index,df['country_txt'].value_counts()[:15].values)
plt.title('Top Countries Affected',fontsize=16)
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation= 90)
plt.show()
```

# The most used weapon

```python
df['weaptype1_txt'].value_counts().sort_values(ascending=False)[:4]
# plt.figure(figsize=(10,10))
plt.pie(x=df['weaptype1_txt'].value_counts().sort_values(ascending=False)[:5],labels=df['weaptype1_txt'].value_counts().sort_values(ascending=False)[:5].index,rotatelabels=True,autopct='%.2f')
plt.legend()
plt.legend(bbox_to_anchor=(1.02,1), loc='lower right',borderaxespad=0)
plt.show()
```

# Target Audience

```python
data['targtype1_txt'].value_counts()[1:10].plot(kind='pie',figsize=(6,8),autopct='%.2f%%')
```

# Count of cases according to region and targettype of terrorists

```python
#count of cases according to region and targettype of terrorists
plt.figure(figsize=(21,7))
sns.countplot(data=df,x='region_txt',hue='targtype1_txt',saturation=1)
plt.legend(bbox_to_anchor=(1.001,1), loc='lower right',borderaxespad=0)
plt.show()
```
